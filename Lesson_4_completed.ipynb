```python
# Concatenate vertically
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})
df_concat = pd.concat([df1, df2], ignore_index=True)

print("Concatenated DataFrame:")
print(df_concat)

# Merge horizontally with df3
df_merged = pd.concat([df_concat, df3], axis=1)

print("\nMerged DataFrame:")
print(df_merged)
```

```python
# Regular function
def square_number(x):
    return x ** 2

# Lambda function
square_lambda = lambda x: x ** 2

# Test both
print("Regular function:", square_number(3))
print("Lambda function:", square_lambda(4))
```

```python
class EvenNumbers:
    def __init__(self):
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current >= 10:
            raise StopIteration
        self.current += 2
        return self.current

# Test iterator
even_iter = EvenNumbers()
print("Iterator output:", list(even_iter))

def even_generator():
    for i in range(2, 11, 2):
        yield i

# Test generator
print("Generator output:", list(even_generator()))
```

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# Map → square all numbers
squared = list(map(lambda x: x**2, numbers))
print("Squared numbers:", squared)

# Reduce → product of all numbers
product = reduce(lambda x, y: x * y, numbers)
print("Product of numbers:", product)

# Filter → only even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers))
print("Even numbers:", evens)
```

```python
# Create Rectangle class
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

# Create objects
rect1 = Rectangle(5, 3)
rect2 = Rectangle(10, 4)

# Display results
print("Rectangle 1 Area:", rect1.area())
print("Rectangle 1 Perimeter:", rect1.perimeter())

print("Rectangle 2 Area:", rect2.area())
print("Rectangle 2 Perimeter:", rect2.perimeter())
```

```python
df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})

# 1. Average salary by department
avg_salary = df_employees.groupby('Department')['Salary'].mean()
print("Average Salary by Department:")
print(avg_salary)

# 2. Employees with salary > 60000
high_salary = df_employees[df_employees['Salary'] > 60000]['Name']
print("\nEmployees with salary > 60000:")
print(high_salary)

# 3. Add Bonus column (10%)
df_employees['Bonus'] = df_employees['Salary'] * 0.10
print("\nDataFrame with Bonus column:")
print(df_employees)
```
